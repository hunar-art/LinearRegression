```python
def loss_function(m,b,points):
    total_error = 0
    for i in range(points):
        x = points.iloc[i].Hours
        y = points.iloc[i].Scores
        total_error += (y - (m*x + b))**2
    
    return total_error / len(points)    
```

```python
def gradient_descent(m_now,b_now,points,L):
    m_gradient = 0
    b_gradient = 0
    n = len(points)
    for i in range(n):
        x = points.iloc[i].Hours
        y = points.iloc[i].Scores
        
        m_gradient += (-2/n) * (y - (m_now*x + b_gradient)) * x
        
        b_gradient += (-2/n) * (y - (m_now*x + b_gradient))
    
    m = m_now - m_gradient * L
    b = b_now - b_gradient * L
            
    return m,b
```

```python
m = 0
b = 0
L = 0.0001
epochs = 400

for i in range(epochs):
    if i % 50 == 0:
        print(i)
    m,b = gradient_descent(m,b,scores,L)

print(m,b)
minimum = int(scores['Hours'].min())
maximum = int(scores['Hours'].max())

plt.scatter(scores.Hours,scores.Scores,marker='x',c='r',label='actual value')

plt.plot(list(range(minimum,maximum)),[m*x+b for x in range(minimum,maximum)],c='b')
plt.show()
```
